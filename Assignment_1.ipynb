## Question 1

### On a Sunday night, there are four people on a bus. The bus has five stops left before it ends its route. Suppose each person will get off the bus at one of the stops, and will do so randomly

```r
## 1.a How many different ways could the people get off the bus?

Total_Ways = 5^4
Total_Ways
```

```r
## 1.b What is the probability that all four people get off the bus on the first stop?

numerator1b = choose(5,4) * 1/5
numerator1b / Total_Ways
```

```r
## 1.d  What is the probability that exactly three of the four people get off the bus on the same stop?

numerator1d = choose(5,3) * choose(5,1)
numerator1d / Total_Ways
```

## Question 9

###  Contruct the probability distribution of X in RStudio with P(X=x) on the Y-axis and the values of X on the X-axis.


```r
x = 0:20
probx = function(x)
{   
    choose(20, x) * (0.25^x) * (0.75 ^ {20 - x})
}

plot(x,probx(x), xlab = "Values of X", ylab = "P(X = x)", ylim = c(0,0.2), main = "Probability Distribution of X", type = "h", col='blue')
```

```r
#What is P(X<10)?

xvalue = 0:9

sum(probx(xvalue))
```

```r
#What is P(X>4)?
xvalue = 5:20

sum(probx(xvalue))
```

```r
# Compute the expected value E[X] and SD(X), with R Studio.
m1 = sum(probx(x))
m2 = sum(probx(x^2))
Varx = m2 - m1^2

m1
Varx
```
